```python
def oddEvenJumps(arr):
    n = len(arr)
    next_higher, next_lower = [0] * n, [0] * n
    stack = []

    # Sorting indices by values in ascending order
    indices = sorted(range(n), key=lambda i: arr[i])

    # Fill next_higher
    for i in indices:
        while stack and i > stack[-1]:
            next_higher[stack.pop()] = i
        stack.append(i)

    # Clear the stack
    stack = []

    # Sorting indices by values in descending order
    indices.sort(key=lambda i: -arr[i])

    # Fill next_lower
    for i in indices:
        while stack and i > stack[-1]:
            next_lower[stack.pop()] = i
        stack.append(i)

    # Initialize dp arrays
    odd_jump, even_jump = [False] * n, [False] * n
    odd_jump[-1], even_jump[-1] = True, True

    # Dynamic programming to mark good indices
    for i in range(n - 2, -1, -1):
        if next_higher[i]:
            odd_jump[i] = even_jump[next_higher[i]]
        if next_lower[i]:
            even_jump[i] = odd_jump[next_lower[i]]

    # Collect good indices
    good_indices = [i for i in range(n)]
    return good_indices

# Example usage:
arr = [1, 2, 3, 2, 1, 4, 4, 5]
good_indices = oddEvenJumps(arr)
print("Good indices:", good_indices)
```

```python
class Solution:
    def oddEvenJumps(self, arr: List[int]) -> int:
        #next_jump_odd[jump_from] = landing, same w/ even
        next_jump_odd, next_jump_even = self.get_next(arr)
        
        #if you can make it to the end if you land here via even or odd
        #ie. dp[i][0] = 1 if you can make it to the end if you land on i via even jump
        dp = [[0] * len(arr) for i in range(2)]

        #can always make to the end if you start at last index
        dp[0][-1], dp[1][-1] = 1,1
        

        #1 index where you can make it to the end at least
        ans = 1
        
        #go backwards, with second to last
        for i in range(len(arr) - 2, -1, -1):

            #if this index is capable of an odd jump
            if next_jump_odd[i] != -1:
                #if you land here on even, whether or not you can finish
                #is determined by whether your next odd jump can make it to the end
                dp[0][i] = dp[1][next_jump_odd[i]]

            #same as above, but in reverse
            if next_jump_even[i] != -1:
                dp[1][i] = dp[0][next_jump_even[i]]

            #since we start with odd jump, you must land there via even
            #so if you land somewhere via even, it means your 
            #first jump has a path to the end
            ans += dp[0][i]
        print(dp)    
        return ans

            
    def get_next(self, arr):
        #we sort the index by their values
        sorted_index = sorted(range(len(arr)), key= lambda x: arr[x])
        print(sorted_index)
        #we set up an array of size arr
        res_odd = [-1] * len(arr)
        stack = []

        #not that this is getting the values of sorted_index
        for i in sorted_index:
            #if stack and the current index is larger than top of stack
            while stack and i > stack[-1]:
                #if there are any previous indexes in the stack
                #that has an index that is smaller than the current index
                #the current index must be the stack's next jump
                #since this list is sorted, this current index's value will be the
                #lowest that is still greater than the popped value
                #also it will be to the right of the popped value since index is larger
                res_odd[stack.pop()] = i
            #we add the current index to stack
            stack.append(i)
        
        #we re-sort the arr, but using negative for smaller
        #we can't just reverse, since we have numbers that could be equal value
        #and we want to sort by their index's second
        sorted_index = sorted(range(len(arr)), key= lambda x: -arr[x])

        #same thing as above, but for smaller than
        stack = []
        res_even = [-1] * len(sorted_index)
        for i in sorted_index:
            while stack and i > stack[-1]:
                res_even[stack.pop()] = i
            stack.append(i)

        return res_odd, res_even
            
```

```python
def splitArray( nums, k: int) -> int:

    dp={}

    def dfs(i,m):
        if m==1:
            return sum(nums[i:])
        if (i,m) in dp:
            return dp[(i,m)]

        res,curSum=float("inf"),0
        for j in range(i, len(nums)-m+1):
            curSum+=nums[j]
            maxSum=max(curSum, dfs(j+1,m-1))
            res=min(res, maxSum)

            if curSum>res:
                break
        dp[(i,m)]=res
        return res

    return dfs(0,k)        
```

```python
from collections import deque

class Node:
    def __init__(self, val, index, neighbour):
        self.val = val
        self.index = index
        self.neighbour = neighbour

    def setValue(self, value):
        self.val = value

def solve(graph):
    queue = deque()
    visited = [False] * len(graph)
    
    for node in graph:
        if node.val == 16:
            queue.append(node)
    
    while queue:
        size = len(queue)
        for _ in range(size):
            node = queue.popleft()
            if visited[node.index]:
                continue
            visited[node.index] = True
            for neighbour_index in node.neighbour:
                neighbour_node = graph[neighbour_index]
                if not visited[neighbour_node.index]:
                    if neighbour_node.val < node.val:
                        neighbour_node.setValue(node.val - 1)
                    queue.append(neighbour_node)
    
    return graph

def main():
    nodes = [
        Node(16, 0, [1, 3]),
        Node(0, 1, [0, 2, 5]),
        Node(0, 2, [1, 4, 6]),
        Node(16, 3, [0, 7]),
        Node(0, 4, [2]),
        Node(0, 5, [1]),
        Node(0, 6, [2]),
        Node(16, 7, [3])
    ]
    
    solved_graph = solve(nodes)
    for node in solved_graph:
        print(f"Node {node.index}: {node.val}")

if __name__ == "__main__":
    main()
```

```python
import torch
import torch.nn.functional as F

class MultiHeadAttention(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, num_heads, dropout_rate=0.1):
        super(MultiHeadAttention, self).__init__()
        assert hidden_dim % num_heads == 0
        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.num_heads = num_heads
        self.head_dim = hidden_dim // num_heads
        
        # Linear projections for queries, keys, and values for each head
        self.Wq = torch.nn.Linear(input_dim, hidden_dim)
        self.Wk = torch.nn.Linear(input_dim, hidden_dim)
        self.Wv = torch.nn.Linear(input_dim, hidden_dim)
        
        # Linear projection for combining attention heads
        self.Wo = torch.nn.Linear(hidden_dim, hidden_dim)
        
        self.dropout = torch.nn.Dropout(dropout_rate)
        
    def forward(self, query, key, value, mask=None):
        batch_size = query.size(0)
        
        # Linear projections for queries, keys, and values for each head
        Q = self.Wq(query).view(batch_size, -1, self.num_heads, self.head_dim)
        K = self.Wk(key).view(batch_size, -1, self.num_heads, self.head_dim)
        V = self.Wv(value).view(batch_size, -1, self.num_heads, self.head_dim)
        
        # Transpose to prepare for attention computation
        Q = Q.permute(0, 2, 1, 3)  # [batch_size, num_heads, seq_len, head_dim]
        K = K.permute(0, 2, 1, 3)  # [batch_size, num_heads, seq_len, head_dim]
        V = V.permute(0, 2, 1, 3)  # [batch_size, num_heads, seq_len, head_dim]
        
        # Compute scaled dot-product attention
        scores = torch.matmul(Q, K.permute(0, 1, 3, 2)) / torch.sqrt(torch.tensor(self.head_dim, dtype=torch.float32))
        
        if mask is not None:
            scores.masked_fill_(mask == 0, float('-inf'))  # Apply mask
        
        attention_weights = F.softmax(scores, dim=-1)
        attention_weights = self.dropout(attention_weights)
        
        # Apply attention to values
        attention_output = torch.matmul(attention_weights, V)
        
        # Transpose and reshape to combine attention heads
        attention_output = attention_output.permute(0, 2, 1, 3).contiguous()  # [batch_size, seq_len, num_heads, head_dim]
        attention_output = attention_output.view(batch_size, -1, self.hidden_dim)
        
        # Linear projection to combine attention heads
        output = self.Wo(attention_output)
        
        return output, attention_weights
```

```python
from typing import List
class Solution:
    def isSolvable(self, words: List[str], result: str) -> bool:
        if max(map(len, words)) > len(result): return False # edge case 
        
        words.append(result)
        digits = [0]*10 
        mp = {} # mapping from letter to digit 
        
        def fn(i, j, val): 
            """Find proper mapping for words[i][~j] and result[~j] via backtracking."""
            if j == len(result): return val == 0 # base condition 
            if i == len(words): return val % 10 == 0 and fn(0, j+1, val//10)
            
            if j >= len(words[i]): return fn(i+1, j, val)
            if words[i][~j] in mp: 
                if j and j+1 == len(words[i]) and mp[words[i][~j]] == 0: return # backtrack (no leading 0)
                if i+1 == len(words): return fn(i+1, j, val - mp[words[i][~j]])
                else: return fn(i+1, j, val + mp[words[i][~j]])
            else: 
                for k, x in enumerate(digits): 
                    if not x and (k or j == 0 or j+1 < len(words[i])): 
                        mp[words[i][~j]] = k
                        digits[k] = 1
                        if i+1 == len(words) and fn(i+1, j, val-k): return True 
                        if i+1 < len(words) and fn(i+1, j, val+k): return True 
                        digits[k] = 0
                        mp.pop(words[i][~j])
        
        return fn(0, 0, 0)
```

```python
from collections import defaultdict


class server_replace:
    def rep_server(self,init_server,rep_server,new_server):
        
        total_requests=sum(init_server)
        
        dict_vals=defaultdict(int)
        
        for i in init_server:
            dict_vals[i]+=1
            
        ans=[]
        print(dict_vals)
        for r,n in zip(rep_server,new_server):
            
            cnt=dict_vals[r]
            
            old_val=cnt*r
            
            dict_vals.pop(r)
            
            dict_vals[n]=cnt
            
            print(total_requests,dict_vals)
            total_requests=total_requests-old_val+n*cnt
            ans.append(total_requests)
            
        return ans
            
```
